```python
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread('img2.jpg')
```

```python
plt.imshow(img)
plt.show()
```

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

edges = cv2.Canny(gray, 100, 200)

bl = cv2.bilateralFilter(img, 10, 75, 75)

cartoon = cv2.bitwise_and(bl, bl, mask=edges)
cartoon = cv2.cvtColor(cartoon, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(cartoon)
plt.show()
```
